```python
def time_between_first_last_reviews(first_rev, last_rev):
    diffs = []
    for i in range(len(first_rev)):
        if type(first_rev[i]) == str and type(last_rev[i]) == str:
            first = [int(x) for x in first_rev[i].split("-")]
            d_first = datetime.datetime(first[0],first[1],first[2])
            last = [int(x) for x in last_rev[i].split("-")]
            d_last = datetime.datetime(last[0],last[1],last[2])
            delta = d_last-d_first
            diffs.append(delta.days)
        else:
            diffs.append(0)
    return diffs

def host_sincer(first_rev):
    if type(first_rev) == str:
        first = [int(x) for x in first_rev.split("-")]
        d_first = datetime.datetime(first[0],first[1],first[2])
        delta = datetime.datetime.now()-d_first
        return delta.days
boom = ['2020-07-09' for i in range(33538)]
df['tbetrev'] = time_between_first_last_reviews(df['first_review'],df['last_review'])
df['hostsincenum'] = df['host_since'].apply(lambda x: host_sincer(x))
```

```python
def zipcode(z):
    if type(z) == int or type(z) == float:
        if type(z) == int:
            return z
        elif z == float('nan'):
            return z
        elif type(z) == float:
            return z
    elif type(z) == str:
        if len(z) == 5:
            return int(z)
        elif len(z) == 7:
            return int(z[:-2])
        elif len(z) == 10:
            return int(z[:-5])
        elif z == '-- default zip code --':
            return z
df['zipcode'] = df['zipcode'].apply(lambda x: zipcode(x))
```

```python
from math import isnan

def to_float(s):
    try:
        ret = float(s)
    except:
        ret = -1
    if isnan(ret):
        ret = -1
    return ret

def to_int(s):
    try:
        ret = int(s)
    except:
        ret = -1
    return ret

categories = [x for x in list(set(df['neighbourhood_group_cleansed'])) if type(x) is str]
categories1 = [x for x in list(set(df['host_response_time'])) if type(x) is str]
categories2 = [x for x in list(set(df['host_is_superhost'])) if type(x) is str]
categories3 = [x for x in list(set(df['host_neighbourhood'])) if type(x) is str]
categories4 = [x for x in list(set(df['host_has_profile_pic'])) if type(x) is str]
categories5 = [x for x in list(set(df['host_identity_verified'])) if type(x) is str]
categories6 = [x for x in list(set(df['neighbourhood_cleansed'])) if type(x) is str]
categories7 = [x for x in list(set(df['city'])) if type(x) is str]
categories8 = [x for x in list(set(df['state'])) if type(x) is str]
categories9 = [x for x in list(set(df['zipcode'])) if type(x) is str]
categories10 = [x for x in list(set(df['market'])) if type(x) is str]
categories11 = [x for x in list(set(df['country_code'])) if type(x) is str]
categories12 = [x for x in list(set(df['property_type'])) if type(x) is str]
categories13 = [x for x in list(set(df['room_type'])) if type(x) is str]
categories14 = [x for x in list(set(df['bed_type'])) if type(x) is str]
categories15 = [x for x in list(set(df['instant_bookable'])) if type(x) is str]
categories16 = [x for x in list(set(df['cancellation_policy'])) if type(x) is str]
categories17 = [x for x in list(set(df['require_guest_profile_picture'])) if type(x) is str]
categories18 = [x for x in list(set(df['require_guest_phone_verification'])) if type(x) is str]
#categories19 = [x for x in list(set(df['name'])) if type(x) is str]
#categories20 = [x for x in list(set(df['summary'])) if type(x) is str]
#categories21 = [x for x in list(set(df['space'])) if type(x) is str]
#categories22 = [x for x in list(set(df['description'])) if type(x) is str]
#categories23 = [x for x in list(set(df['neighborhood_overview'])) if type(x) is str]
#categories24 = [x for x in list(set(df['notes'])) if type(x) is str]
#categories25 = [x for x in list(set(df['transit'])) if type(x) is str]
#categories26 = [x for x in list(set(df['access'])) if type(x) is str]
#categories27 = [x for x in list(set(df['interaction'])) if type(x) is str]
#categories28 = [x for x in list(set(df['host_id'])) if type(x) is str]
#categories29 = [x for x in list(set(df['host_name'])) if type(x) is str]
#categories30 = [x for x in list(set(df['host_since'])) if type(x) is str]
#categories31 = [x for x in list(set(df['host_location'])) if type(x) is str]
#categories32 = [x for x in list(set(df['host_about'])) if type(x) is str]


#categories33 = [x for x in list(set(df['first_review'])) if type(x) is str]
#categories34 = [x for x in list(set(df['last_review'])) if type(x) is str]


#Feature Engineering
def create_feature(row):
    guest_num = to_int(row.guests_included)
    #guest_cap = to_int(row.extra_people)
    review = to_float(row.review_scores_rating)
    #mine
    reviews_per_month = to_float(row.reviews_per_month)
    calculated_host_listings_count = to_int(row.calculated_host_listings_count)
    number_of_reviews = to_int(row.number_of_reviews)
    maximum_nights = to_int(row.maximum_nights)
    minimum_nights = to_int(row.minimum_nights)
    extra_people = to_int(row.extra_people[1:][:-3])
    guests_included = to_int(row.guests_included)
    square_feet = to_float(row.square_feet)
    beds = to_int(row.beds)
    bedrooms = to_int(row.bedrooms)
    bathrooms = to_int(row.bathrooms)
    accomodates = to_int(row.accommodates)
    host_listings_count = to_float(row.host_listings_count)
    host_response_rate = to_int(row['host_response_rate'])
    numver = to_int(row['numver'])
    amenlen = to_int(row['amenlen'])
    review1 = to_float(row.review_scores_accuracy)
    review2 = to_float(row.review_scores_cleanliness)
    review3 = to_float(row.review_scores_checkin)
    review4 = to_float(row.review_scores_communication)
    review5 = to_float(row.review_scores_location)
    review6 = to_float(row.review_scores_value)
    tbetrev = to_int(row.tbetrev)
    host_since_num = to_float(row.hostsincenum)
    
    
    one_hot = [int(row.neighbourhood_group_cleansed == category) for category in categories]
    #mine
    one_hot1 = [int(row.host_response_time == category) for category in categories1]
    one_hot2 = [int(row.host_is_superhost == category) for category in categories2]
    one_hot3 = [int(row.host_neighbourhood == category) for category in categories3] 
    one_hot4 = [int(row.host_has_profile_pic == category) for category in categories4] 
    one_hot5 = [int(row.host_identity_verified == category) for category in categories5]
    one_hot6 = [int(row.neighbourhood_cleansed == category) for category in categories6] 
    one_hot7 = [int(row.city == category) for category in categories7] 
    one_hot8 = [int(row.state == category) for category in categories8] 
    one_hot9 = [int(row.zipcode == category) for category in categories9] 
    one_hot10 = [int(row.market == category) for category in categories10] 
    one_hot11 = [int(row.country_code == category) for category in categories11] 
    one_hot12 = [int(row.property_type == category) for category in categories12]
    one_hot13 = [int(row.room_type == category) for category in categories13]
    one_hot14 = [int(row.bed_type == category) for category in categories14]
    one_hot15 = [int(row.instant_bookable == category) for category in categories15]
    one_hot16 = [int(row.cancellation_policy == category) for category in categories16]
    one_hot17 = [int(row.require_guest_profile_picture == category) for category in categories17]
    one_hot18 = [int(row.require_guest_phone_verification == category) for category in categories18]
    #one_hot33 = [int(row.first_review == category) for category in categories33]
    #one_hot34 = [int(row.last_review == category) for category in categories34]
    #one_hot35 = [int(row.room_type == category) for category in categories35]
    #one_hot36 = [int(row.room_type == category) for category in categories36]
    #one_hot37 = [int(row.room_type == category) for category in categories37]
    #one_hot38 = [int(row.room_type == category) for category in categories38]
    
    
    
    
    
    return [guest_num, bedrooms, bathrooms, review,
            reviews_per_month,calculated_host_listings_count,
            number_of_reviews,maximum_nights,minimum_nights,
            extra_people,guests_included,square_feet,beds,bedrooms, 
            bathrooms, accomodates, host_listings_count,host_response_rate,numver,amenlen
            ,review1,review2,review3,review4,review5,review6,tbetrev,host_since_num
            ] + one_hot + one_hot1 + one_hot2 + one_hot3 + one_hot4 + one_hot5 + one_hot6 + one_hot7 + one_hot8 + one_hot9 + one_hot10 + one_hot11 + one_hot12 + one_hot13 + one_hot14 + one_hot15 + one_hot16 + one_hot17 + one_hot18 



train_X, train_y = [], []
for (idx, row) in df.iterrows():
    price = float(row.price)
    feature = create_feature(row)
    train_X.append(feature)
    train_y.append(price)

print(len(train_X), len(train_y))
```
